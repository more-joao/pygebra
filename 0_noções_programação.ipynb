# **Fundamentos de Programação**

Este capítulo tem o intento de fornecer uma visão (resumidamente) anatômica do processo de programação. Para esse desígnio, percorramos os estratos que o compõem.
<br><br>

### **1 Computadores**

Uma precisa definição para computador é: máquina programável capaz de receber, armazenar e manipular dados.
<br>

Esses dois predicados bem delimitam o conjunto de máquinas que podem ser consideradas computadores, assim como determinam a arquitetura e organização comum dos componentes desses dispositivos. Frisa-se: computadores devem ser **programáveis** e capazes de **processar dados**.

Discutamos acerca das duas mais importantes partes de um sistema de computação.<br><br>

#### **1.1 Memória Principal**

A memória principal é dita volátil; isto é, não é persistente para além do tempo de execução dos programas e da existência de energia. Durante a execução de programas pelo processador, é ela quem recebe e retorna os dados pertinentes.

Em contraste com a memória secundária (que é persistente e não volátil), a memória principal é rápida de sorte a acompanhar a velocidade do processador.

Cada local na memória principal possui um **endereço único**. Destarte, todo dado nela residente pode ser acessado mediante o endereço de onde figura.
<br><br>
#### **1.2 Processador**

O processador é o componente principal de um sistema de computação. É o processador que traz ao computador o aspecto de **ser programável**.

Nele são realizados cálculos lógicos e aritméticos, envio e busca de dados na memória e armazenamento temporário de resultados (em registradores [componentes de memória ainda mais rápidos que a memória principal]).

Todo o processamento decorre fundamentalmente mediante sinais elétricos que percorrem caminhos chamados barramentos. Esses sinais são interpretados pelo processador como cadeias binárias e vice-versa.

Cada processador possui um conjunto de instruções básicas predeterminadas a partir das quais deve traduzir quaisquer outras instruções.
