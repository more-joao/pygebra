# **Fundamentos de Programação**

Este capítulo tem o intento de fornecer uma visão (resumidamente) anatômica do processo de programação. Para esse desígnio, percorramos os estratos que o compõem.
<br><br>

### **1 Computadores**

Uma precisa definição para computador é: máquina programável capaz de receber, armazenar e manipular dados.
<br>

Esses dois predicados bem delimitam o conjunto de máquinas que podem ser consideradas computadores, assim como determinam a arquitetura e organização comum dos componentes desses dispositivos. Frisa-se: computadores devem ser **programáveis** e capazes de **processar dados**.

Ante esse propósito, foi adotada a arquitetura de von Neumann. Discutamos acerca das duas mais importantes partes de um sistema de computação para ela.<br><br>

#### **1.1 Memória Principal**

A memória principal é dita volátil; isto é, não é persistente para além do tempo de execução dos programas e da existência de energia. Durante a execução de programas pelo processador, é ela quem recebe e retorna os dados pertinentes.

Cada local na memória principal possui um **endereço único**. Destarte, todo dado nela residente pode ser acessado mediante o endereço de onde figura.
<br><br>
#### **1.2 Processador**

O processador é o componente principal de um sistema de computação. É o processador lhe traz o aspecto de **programabilidade**.

Nele são realizados cálculos lógicos e aritméticos, envio e busca de dados na memória e armazenamento temporário de resultados (*registradores*).

Todo o processamento decorre fundamentalmente mediante sinais elétricos que percorrem caminhos chamados barramentos. Esses sinais são interpretados pelo processador como cadeias binárias e vice-versa.

Cada processador possui um **conjunto de instruções básicas** predeterminadas a partir das quais deve traduzir quaisquer outras instruções.
<br><br>

### **2 Programas de Computador**

Programas de computador são conjuntos de instruções que, como um todo, visam desempenhar uma tarefa específica. Todo programa de computador possui um **código-fonte**, que é escrito em uma linguagem de programação.
<br><br>
#### **2.1 Linguagem de Programação**

Existe uma miríade de linguagens de programação, que são sistemas de palavras regidas por regras de lógica e de sintaxe próprias. Para toda linguagem de programação, há um programa particular que realiza a tradução do código-fonte para *código de máquina* (binário). Interpretando as cadeias binárias como sequências de instruções básicas, o processador realiza as imperativas do programa de computador.

Esses programas tradutores podem ser **compiladores** ou **interpretadores**. Compiladores traduzem o código-fonte por completo antes da execução, enquanto interpretadores traduzem e executam comando por comando.

Linguagens de programação também são classificadas em **alto nível** e **baixo nível**. Aspectos como semelhança ao vocabulário humano, independência de hardware e gerenciamento de memória são atribuídos à níveis mais altos. Vejamos uma comparação.
